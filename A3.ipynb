```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros iniciales
n_neutrons = 1000  # Número inicial de neutrones
energy_initial = np.random.uniform(0.1, 10, n_neutrons)  # Energías iniciales en MeV (rápidos)

# Moderación de neutrones
def moderate_neutrons(energies, factor=0.1):
    """
    Reduce la energía de los neutrones para simular la moderación.
    factor: fracción de energía que conservan los neutrones tras moderarse.
    """
    return energies * factor

# Probabilidad de captura
def capture_probability(energy, cross_section=0.5):
    """
    Calcula la probabilidad de captura basada en la energía.
    cross_section: parámetro que determina la probabilidad de interacción (sección eficaz).
    """
    return np.exp(-energy / cross_section)

# Simular moderación y captura
moderated_energies = moderate_neutrons(energy_initial)
capture_probabilities = capture_probability(moderated_energies)

# Resultados: determinar cuántos neutrones son capturados
captured = np.random.rand(n_neutrons) < capture_probabilities
captured_count = np.sum(captured)

# Visualización
plt.figure(figsize=(10, 6))
plt.hist(energy_initial, bins=30, alpha=0.5, label="Energías iniciales (MeV)")
plt.hist(moderated_energies, bins=30, alpha=0.5, label="Energías moderadas (MeV)")
plt.title("Energías de neutrones antes y después de moderación")
plt.xlabel("Energía (MeV)")
plt.ylabel("Frecuencia")
plt.legend()
plt.grid()
plt.show()

print(f"Neutrones capturados: {captured_count}/{n_neutrons}")
```
